**Switch in R**

--------
A switch statement chooses one of several options to run based on the value of an input. It’s like a menu where you pick a number or name, and the program performs the corresponding task.

```
switch(expression, option1, option2, ...)
```

````r
a <- 3

res <- switch(a,
"Hello",
"Hii",
"Fine",
"Go"
)
res
````

````r
# Example of switch statement
option <- 3

result <- switch(option,
  "Option 1 chosen",   # When option is 1
  "Option 2 chosen",   # When option is 2
  "Option 3 chosen"    # When option is 3
)

print(result)  # Output: "Option 2 chosen"
````

````r
choice <- "C"

result <- switch(choice,
  A = "You chose A",
  B = "You chose B",
  C = "You chose C"
)

print(result)  # Output: "You chose B"
````

### **Loops in R**
-----------
Loops are used to repeat a block of code multiple times. They are helpful when you need to perform the same operation for multiple values or until a specific condition is met.

*R has two loop commands:*

1. **while** loops
2. **for** loops

**While Loop**

--------------
A while loop keeps running as long as the condition is true.

```
while(conditions){
  statements
}
```

````r
3 < 6
````

````r
a <- 2

while(a < 6){
  print("Hello, How are you?")
  a <- a + 1
}

print("Hello")
````

````r
# while loop
i <- 1
while (i < 6) {
  print(i)
  i <- i + 1
}
````

**For Loop**

--------------
With the for loop we can execute a set of statements, once for each item in a vector, array, list, etc..

```
for(expressions){
  statements
}
```

````r
x <- 1:5
print(x)
````

````r
for(a in x){
  print(a)
}
````

````r
a <- c("Janak", "Hari", "Shyam")

for(name in a){
  print(name)
}
````

````r
# for loop
for (i in 1:5){
  print("NO")
  break #Jump/exit from loop
}
````

**nested** **loop**

````r
i <- 0
while (i < 6) {
  i <- i + 1
  if (i == 3) {
    next #skip upcoming expression without terminating the loop
  }
  print(i)
}
````

**repeat Loop**

A repeat loop runs until you use the break statement to exit it

------------
```
repeat{
  statements
  if(exp.){
    break
  }
}
```

````r
j <- 1

repeat{
  print("Hii")
  j <- j+1
  if(j > 3){
    break
  }
}
````

````r
# Example: Count to 5 using repeat
x <- 1
repeat {
  print(x)
  x <- x + 1
  if (x > 5) {
    break
  }
}
````

3. **Correlation and Covariance**

Correlation measures the linear relationship between two variables. Covariance measures how two variables vary together.

````r
# Sample Data
x <- c(1, 2, 3, 4, 5)
y <- c(2, 4, 6, 8, 10)

# Covariance
cov_xy <- cov(x, y)
print(paste("Covariance:", cov_xy))

# Correlation
cor_xy <- cor(x, y)
print(paste("Correlation:", cor_xy))
````

4. **Hypothesis Testing**

Perform a t-test to compare the means of two samples.

````r
# Two Samples
sample1 <- c(5, 6, 7, 8, 9)
sample2 <- c(6, 7, 8, 9, 10)

# t-test
t_test_result <- t.test(sample1, sample2, alternative = "two.sided")
print(t_test_result)
````

5. **Linear Regression**

Linear regression models the relationship between a dependent variable and one or more independent variables.

````r
# Sample Data
data <- data.frame(
  x = c(1, 2, 3, 4, 5),
  y = c(2, 3, 5, 7, 11)
)

# Linear Model
model <- lm(y ~ x, data = data)
# ~: Indicates a relationship between the dependent and independent variables.
# Summary of Model
print(summary(model))

# Plot the Data and Regression Line
plot(data$x, data$y, main="Linear Regression", xlab="X", ylab="Y", pch=16)
abline(model, col="blue")
````

**Layout SetUp**

**Using par(mfrow)**

The par() function with the mfrow argument divides the plotting area into a matrix of subplots.

````r
# Set up a 2x2 plotting layout
par(mfrow = c(2, 2))

# Plot 1: Histogram
set.seed(42)
data1 <- rnorm(100)
hist(data1, main = "Histogram", col = "lightblue")

# Plot 2: Scatter Plot
x <- rnorm(100)
y <- 2 * x + rnorm(100)
plot(x, y, main = "Scatter Plot", col = "darkgreen", pch = 19)

# Plot 3: Box Plot
data2 <- rnorm(100)
boxplot(data2, main = "Boxplot", col = "pink")

# Plot 4: Bar Plot
categories <- c("A", "B", "C")
values <- c(10, 20, 15)
barplot(values, names.arg = categories, main = "Bar Plot", col = "lightgray")

````

**Using layout()**

The layout() function provides more control over the arrangement of plots.

````r
# Define layout: 3 plots with 2 rows (first row spans two columns)
layout(matrix(c(1, 1, 2, 3, 4, 4), nrow = 3, byrow = TRUE))

# Plot 1: Larger Plot Spanning Two Columns
hist(data1, main = "Histogram (Spans Two Columns)", col = "lightblue")

# Plot 2: Scatter Plot
plot(x, y, main = "Scatter Plot", col = "darkgreen", pch = 19)

# Plot 3: Box Plot
boxplot(data2, main = "Boxplot", col = "pink")

# Plot 4: Larger Plot Spanning Two Columns
hist(data1, main = "Histogram (Spans Two Columns)", col = "lightblue")
````

**Using par(mfg) for Custom Placement**

The par(mfg) argument allows manual placement of plots in specific positions in the grid.

````r
# Set up a 2x2 layout
par(mfrow = c(2, 2))

# Place first plot in position (1, 1)
par(mfg = c(1, 1))
plot(x, y, main = "Scatter Plot")

# Place second plot in position (1, 2)
par(mfg = c(1, 2))
hist(data1, main = "Histogram")

# Place third plot in position (2, 1)
par(mfg = c(2, 1))
boxplot(data2, main = "Boxplot")

# Place fourth plot in position (2, 2)
par(mfg = c(2, 2))
barplot(values, names.arg = categories, main = "Bar Plot")
````
